```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
from nltk.corpus import stopwords
from wordcloud import WordCloud
```

```python
df = pd.read_csv("dataset.csv")
df.head()
```

```python
df.isnull().sum()
```

```python
df.dropna(inplace=True)

df.isnull().sum()
```

Word Cloud for Questions

```python
# Combine sentences into a single text
text = " ".join(df['question'])

# Define stopwords to remove common words
stop_words = set(stopwords.words('english'))

# Create a WordCloud object with custom settings
wordcloud = WordCloud(
    background_color='white',
    stopwords=stop_words,
    width=800,
    height=400
).generate(text)

# Display the word cloud
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title('Word Cloud for Questions')
plt.show()
```

Visualizing the length of the questions

```python
# Calculate the length of each sentence
df['question_Length'] = df['question'].apply(len)

# Create a histogram of sentence lengths
plt.hist(df['question_Length'], bins=10, edgecolor='k')
plt.xlabel('Questions Length')
plt.ylabel('Frequency')
plt.title('Histogram of Question Lengths')
plt.show()
```
